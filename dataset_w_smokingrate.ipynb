```python
import pandas as pd
```

```python
smoking_data = 'Resources/adults-smoking-2007-2018.csv'
smoking_df = pd.read_csv(smoking_data)
smoking_df.head()
```

```python
smoking_df = smoking_df.drop(columns=['Entity', 'Continent', 'Prevalence of current tobacco use (% of adults).1', 'Year.1'])
```

```python
smoking_df = smoking_df.rename(columns={"Prevalence of current tobacco use (% of adults)": 'smoking_prevalence'})
```

```python
smoking_df
```

```python
merge_id = []

for i, row in smoking_df.iterrows():
    merge_id.append(f"{smoking_df['Code'][i]}_{smoking_df['Year'][i]}")

smoking_df['merge_id'] = merge_id
```

```python
smoking_df
```

```python
data = 'Resources/merged_data_w_landarea.csv'
df = pd.read_csv(data)
df
```

```python
merged_df = pd.merge(df, smoking_df, left_on='merge_id', right_on='merge_id', how='outer')
merged_df
```

```python
merged_df.to_csv('Resources/co2_deaths_cigs_land_smoking.csv')
```
