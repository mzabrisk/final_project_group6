```python
import pandas as pd
```

```python
data = 'Resources/DataWMinGPD.csv'
df = pd.read_csv(data)
df = df.drop(columns='Unnamed: 0')
```

```python
data2 = 'Resources/country_pop_by_year.csv'
df2 = pd.read_csv(data2)
df2
```

```python
years = list(range(1960,2022))
country_years = []

for country in df2['Country Code']:
    for year in years:
        country_years.append(f'{country}_{year}')

country_years
```

```python
pop_df = pd.DataFrame(columns=['merge_id', 'population'])
pop_df['merge_id'] = country_years
pop_df
```

```python
column = [str(x) for x in years]
population = []

for i, row in df2.iterrows():
    for x in column:
        population.append(df2[x][i])
```

```python
pop_df['population'] = population
pop_df
```

```python
merged_df = pd.merge(df,pop_df, left_on='merge_id', right_on='merge_id')
merged_df
```

```python
merged_df.to_csv('Resources/DataWMinGPDPop.csv', index=False)
```
