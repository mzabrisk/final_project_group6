```python
# Initial imports
import pandas as pd
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
file_path = "../Resources/DataWAvgGdp.csv"
merged_df = pd.read_csv(file_path, index_col="country_code")
merged_df.head(10)
```

```python
merged_df.isnull().sum()
```

```python
country_name=pd.DataFrame(merged_df['country_name'])
country_name
```

```python
merged_df.drop(columns="merge_id", inplace=True)
merged_df
```

```python
merged_df.dtypes
```

```python
x = pd.get_dummies(merged_df, columns=['country_name'])
x
```

```python
x_scaled =  StandardScaler().fit_transform(x)
x_scaled
```

```python
pca = PCA(n_components=2)
scaled_pca = pca.fit_transform(x_scaled)
```

```python
pcs_df = pd.DataFrame(
    data=scaled_pca, columns=["PC 1", "PC 2"], index=merged_df.index
)
pcs_df.head()
```

```python
inertia = []
k = list(range(1, 11))


for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=, random_state=0)
# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
clustered_df = pd.concat([merged_df, pcs_df], axis="columns")

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
clustered_df = pd.concat([clustered_df, country_name], axis="columns")

#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df["Class"] = model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(
    clustered_df,
    x="PC 2",
    y="PC 1",
    color="Class",
    symbol="Class",
   
)
fig.update_layout(legend = dict(x=0, y=1))
fig.show()
```
