```python
import pandas as pd
```

```python
co2_data = 'Resources/co2_emissions_kt_by_country.csv'
co2_df = pd.read_csv(co2_data)
print(co2_df.shape)
co2_df.head()
```

```python
print(len(co2_df['country_code'].value_counts()))
print(co2_df['country_code'].value_counts())
```

```python
# creating new index for co2_df to enable merging dfs
co2_index = []

for i, row in co2_df.iterrows():
    value = f"{co2_df['country_code'][i]}_{co2_df['year'][i]}"
    co2_index.append(value)

co2_index
```

```python
# setting a new index for co2_df
co2_df['new_index'] = co2_index
co2_df
```

```python
lungcancer_data = 'Resources/lung-cancer-deaths-per-100000-by-sex-1950-2002.csv'
lungcancer_df = pd.read_csv(lungcancer_data)
print(lungcancer_df.shape)
lungcancer_df.head()
```

```python
print(len(lungcancer_df['Code'].value_counts()))
print(lungcancer_df['Code'].value_counts())
```

```python
# creating new index for lungcancer_df to enable merging dfs
lungcancer_index = []

for i, row in lungcancer_df.iterrows():
    value = f"{lungcancer_df['Code'][i]}_{lungcancer_df['Year'][i]}"
    lungcancer_index.append(value)

lungcancer_index
```

```python
cigarettes_data = 'Resources/sales-of-cigarettes-per-adult-per-day.csv'
cigarettes_df = pd.read_csv(cigarettes_data)
print(cigarettes_df.shape)
cigarettes_df.head()
```

```python
print(len(cigarettes_df['Code'].value_counts()))
print(cigarettes_df['Code'].value_counts())
```

```python
# creating new index for cigarettes_df to enable merging dfs
cigarettes_index = []

for i, row in cigarettes_df.iterrows():
    value = f"{cigarettes_df['Code'][i]}_{cigarettes_df['Year'][i]}"
    cigarettes_index.append(value)

cigarettes_index
```

```python
# setting a new index for cigarettes_df
cigarettes_df['new_index'] = cigarettes_index
cigarettes_df
```

```python
# merging dataframes

merged_df1 = pd.merge(left= co2_df, right= lungcancer_df, left_on='new_index', right_on='new_index')
merged_df1
```

```python
merged_df = pd.merge(left=merged_df1, right=cigarettes_df, left_on='new_index', right_on='new_index')
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns=['Entity_x', 'Code_x', 'Year_x', 'Entity_y', 'Code_y', 'Year_y'])
```

```python
merged_df.rename(columns={'value': 'co2_emissions', 'new_index': 'merge_id'}, inplace=True)
merged_df
```
