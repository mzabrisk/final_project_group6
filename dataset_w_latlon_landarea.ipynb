```python
import pandas as pd
```

```python
data = 'Resources/merged_data.csv'
df = pd.read_csv(data)
df
```

```python
land_data = 'Resources/land_area.csv'
land_df = pd.read_csv(land_data)
land_df_condensed =pd.DataFrame()

land_df_condensed[['Country Code', 'land_area']] = land_df[['Country Code', '2020']]
land_df_condensed.head()
```

```python
merged_data_w_land_df = pd.merge(df, land_df_condensed, left_on='country_code', right_on='Country Code')
merged_data_w_land_df.drop(columns=['Country Code'], inplace = True)
```

```python
merged_data_w_land_df.to_csv('Resources/merged_data_w_landarea.csv', index=False)
```
