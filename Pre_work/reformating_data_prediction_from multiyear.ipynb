```python
import pandas as pd
```

```python
data = 'Resources/DataWMinGPDPop.csv'
df = pd.read_csv(data)
df
```

```python
df['country_code'].value_counts()
```

```python
df['gdp_normalized'] = df['GDP'] / df['population']
df['co2_normalized'] = df['co2_emissions'] / df['land_area']
df = df.drop(columns=['GDP', 'population', 'co2_emissions', 'land_area'])
df.head()
```

```python
# creating df that uses previous 8 years to predict deaths
columns = ['country_code', 'country_name', 'year', 'co2_normalized', 'merge_id', 'male_death_rate', 'female_death_rate', 'cigarette_sales', 'gdp_normalized', 'co2_emissions_-8','co2_emissions_-7', 'co2_emissions_-6','co2_emissions_-5', 'co2_emissions_-4', 'co2_emissions_-3', 'co2_emissions_-2', 'co2_emissions_-1', 'cigarette_sales_-8', 'cigarette_sales_-7', 'cigarette_sales_-6', 'cigarette_sales_-5', 'cigarette_sales_-4', 'cigarette_sales_-3', 'cigarette_sales_-2', 'cigarette_sales_-1', 'GDP_-8', 'GDP_-7', 'GDP_-6', 'GDP_-5', 'GDP_-4', 'GDP_-3', 'GDP_-2', 'GDP_-1'] 

multiyear_df = pd.DataFrame(columns=columns)
multiyear_df
```

```python
multiyear_df['country_code'] = df['country_code']
multiyear_df['country_name'] = df['country_name']
multiyear_df['year'] = df['year']
multiyear_df['co2_normalized'] = df['co2_normalized']
multiyear_df['merge_id'] = df['merge_id']
multiyear_df['male_death_rate'] = df['male_death_rate']
multiyear_df['female_death_rate'] = df['female_death_rate']
multiyear_df['cigarette_sales'] = df['cigarette_sales']
multiyear_df['gdp_normalized'] = df['gdp_normalized']
multiyear_df.head()
```

```python
year_past = [8,7,6,5,4,3,2,1,]

for j in year_past: 
    for i, row in multiyear_df[j:].iterrows():
        if (multiyear_df['country_code'][i] == df['country_code'][i-j]) & ((multiyear_df['year'][i]) == (df['year'][i])):
            multiyear_df[f'co2_emissions_-{j}'][i] = (df['co2_normalized'][i-j])
            multiyear_df[f'cigarette_sales_-{j}'][i] = (df['cigarette_sales'][i-j])
            multiyear_df[f'GDP_-{j}'][i] = (df['gdp_normalized'][i-j])

multiyear_df.head(20)
```

```python
multiyear_df.isna().sum()
```

```python
multiyear_df = multiyear_df.dropna()

multiyear_df
```

```python
multiyear_df.to_csv('Resources/eight_year_cycle.csv', index=False)
```
