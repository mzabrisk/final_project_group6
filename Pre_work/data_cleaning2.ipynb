```python
import pandas as pd
```

```python
data = 'Resources/DataWMinGPDPop.csv'
df = pd.read_csv(data)
print(df.shape)
df.head()
```

```python
df = df.drop(columns=['country_code', 'merge_id', 'Country Name', 'Country Code', 'Year'])
df.columns
```

```python
df['gdp_norm'] = df['GDP'] / df['population']
df['co2_norm'] = df['co2_emissions'] / df['land_area']
```

```python
# adding 'clear' data
df.loc[len(df.index)] = ['', 1990, 0, 0, 0, 0, 0, 0, 0, 0, 0]
df
```

```python
df = df.fillna(0)
```

```python
# saving csv file
df.to_csv('Resources/cleaned_data_for_json.csv', index=False)
```

```python
country_list = list(df.country_name.unique())
country_list.sort()
country_list
```

```python
# creating json file with correct format

data = []
# countries = list(merged_df.country_name.unique())

for j, country in enumerate(country_list):
    year_list = []
    co2_list = []
    female_death_list = []
    male_death_list = []
    cigarette_sale_list = []
    land_area_list = []
    gdp_list = []
    population_list = []
    gdp_norm_list = []
    co2_norm_list = []
    years = {'country': country_list[j], 'years': year_list, 'co2': co2_list, 'female_dealth_rate': female_death_list, 'male_death_rate': male_death_list, 'cigarette_sales': cigarette_sale_list, 'land_area': land_area_list, 'gdp': gdp_list, 'population': population_list, 'co2_norm': co2_norm_list, 'gdp_norm': gdp_norm_list}

    data.append(years)

    for i, row in df.iterrows():
        if df['country_name'][i] == country_list[j]:
            year_list.append(df['year'][i])
            co2_list.append(df['co2_emissions'][i])
            female_death_list.append(df['female_death_rate'][i])
            male_death_list.append(df['male_death_rate'][i])
            cigarette_sale_list.append(df['cigarette_sales'][i])
            land_area_list.append(df['land_area'][i])
            gdp_list.append(df['GDP'][i])
            population_list.append(df['population'][i])
            gdp_norm_list.append(df['gdp_norm'][i])
            co2_norm_list.append(df['co2_norm'][i])
        

data
```

```python
# combining with country list
output_data = {'country': country_list, 'data': data}
output_data
```

```python
# exporting .json file

import numpy as np
from numpyencoder import NumpyEncoder
import json

with open('dashboard/static/resources/cleaned_json_data.json', 'w') as f:
    json.dump(output_data, f, indent=4, cls=NumpyEncoder)
```
